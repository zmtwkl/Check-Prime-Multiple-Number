# Task 1 - Multiple

**Print out integers which are multiples of 3 from 1 to 22.**

## Using For Loops

```python
for x in range (3, 22, 3) :
  print(x)
```

## Using While Loops

```python
i = 3
while i < 22:
  print(i)
  i = i + 3
```

# Task 2 - Prime Number

```python
def prime_check (x):
  if x > 1:
   # Check if factor exist  
   for i in range(2,x):
       if (x % i) == 0:
           print(x,"is not a Prime number")
           break
   else:
       print(x,"is a Prime number")
       
# Else if the input number is less than or equal to 1
  else:
   print(x,"is not a Prime number")
```
